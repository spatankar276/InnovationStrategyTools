```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt



```

```python
# get google takes in a list of 'keywords' and returns a dataframe with search trends (web trends)
# Note: keywords list must be limited to 5 due to api restrictions
# Note: Data is broken down by regions

def get_google_news(kw_list):
    
    from pytrends.request import TrendReq
    pytrend = TrendReq()
    
    if(len(kw_list) > 5):
        return print('The key words list must have 5 or less key words')
        
    diet_region_dict = {}
    regions = ['ZA', 'US', 'MX', 'IN', 'GB', 'CN', 'CA', 'BR', 'AR', 'AU', 'NZ']
    
    
    for region in regions:
        pytrend.build_payload(kw_list, geo = region, gprop = 'web')
        diet_region_dict[region] = pytrend.interest_over_time()
    
    za = pd.DataFrame(diet_region_dict.get('ZA'))
    us = pd.DataFrame(diet_region_dict.get('US'))
    mx = pd.DataFrame(diet_region_dict.get('MX'))
    ind = pd.DataFrame(diet_region_dict.get('IN'))
    gb = pd.DataFrame(diet_region_dict.get('GB'))
    cn = pd.DataFrame(diet_region_dict.get('CN'))
    ca = pd.DataFrame(diet_region_dict.get('CA'))
    br = pd.DataFrame(diet_region_dict.get('BR'))
    ar = pd.DataFrame(diet_region_dict.get('AR'))
    au = pd.DataFrame(diet_region_dict.get('AU'))
    nz = pd.DataFrame(diet_region_dict.get('NZ'))
    
    za['Region'] = 'za'
    us['Region'] = 'us'
    mx['Region'] = 'mx'
    ind['Region'] = 'in'
    gb['Region'] = 'gb'
    cn['Region'] = 'cn'
    ca['Region'] = 'ca'
    br['Region'] = 'br'
    ar['Region'] = 'ar'
    au['Region'] = 'au'
    nz['Region'] = 'nz'
    
    dfs = [us, mx, ind, gb, cn, ca, br, ar, au, nz]
    
    for df in dfs:
        temp = za.append(df)
        za = temp
        
    return za
        
        
        
```
